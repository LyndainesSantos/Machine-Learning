```python
#import PCA, dataset, matplotlib.pyplot
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
# Carregando o conjunto de dados de dígitos escritos à mão
digits = load_digits()
X = digits.data
y = digits.target
```

```python
# Reduzindo a dimensionalidade para 2 componentes principais usando PCA
# Sintaxe: PCA (n_components = número de componentes)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Visualizando os dados reduzidos em um gráfico de dispersão
plt.figure(figsize=(8, 6))
for i in range(10):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=str(i), alpha=0.6)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.show()
```
